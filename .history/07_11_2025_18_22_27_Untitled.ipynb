```r
library(tidyverse)
library(knitr)
```

```r
players <- read_csv("players.csv") |>
    mutate(experience = factor(experience),
           gender = factor(gender))
sessions <- read_csv("sessions.csv") |>
    mutate(start_time = dmy_hm(start_time, tz = "UTC"),
           end_time = dmy_hm(end_time, tz = "UTC"),
           duration_min = as.numeric(difftime(end_time, start_time, units = "mins")))

glimpse(players)
glimpse(sessions)
```

```r
players_means <- players |>
    summarize(mean_played_hours = mean(played_hours, na.rm = TRUE),
              mean_age = mean(Age, na.rm = TRUE))
players_means
```

```r
proportion_subscribed <- ggplot(players, aes(x = experience, fill = subscribe)) +
    geom_bar(position = "fill") +
    labs(y = "Proportion subscribed")
proportion_subscribed
```

```r
ggplot(players, aes(x = experience, fill = subscribe)) +
    geom_bar(position = "fill") +
    labs(x = "Experience level", y = "Proportion subscribed", title = "Newsletter subscription by experience")
```
